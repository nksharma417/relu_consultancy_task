```python
from selenium import webdriver             ### Importing some libraries which I am gonna use in this 
from selenium.webdriver.common.by import By

import pandas as pd 
import numpy as np 
import warnings
warnings.filterwarnings('ignore')
from bs4 import BeautifulSoup
import time
import random 
import re
import json
```

```python
df=pd.read_csv(r'D:/relu/Amazon Scraping - Sheet1.csv')        ### Reading data from the csv file of products and asin 
url_format='https://www.amazon.{country}/dp/{asin}'
```

```python
def create_amazon_url(row):
    return f"https://www.amazon.{row['country']}/dp/{row['Asin']}"
df['amazon_url'] = df.apply(create_amazon_url, axis=1)         ### Creating a new column which will have complete url 
```

```python
url_list=[i for i in df['amazon_url']]
random_sample = random.sample(url_list, 100)        ### Taking a sample of 100 websites from all the 1000 to do the task
random_sample
```

```python
random_sample_1 = random.sample(url_list, 10)
```

```python
driver=webdriver.Firefox()          ### I will use firefox webdriver
```

```python
dictionary=[]           ### Creating a empty list to store data 
```

#### The below code is written with currently keeping the thing in mind that our scrapper will not interact with any captcha. If captcha appears in between we will modify the code. 

```python
for i in random_sample_1:          ### Writing the code to collect data from every of the 100 urls
    driver.get(i)

    try:
        z=driver.find_element(By.XPATH,'//*[@id="sp-cc-accept"]')
        z.click()
    except:
        pass

    try:
        a=driver.find_element(By.XPATH,'//*[@id="productTitle"]')
        a1=a.text
        b=driver.find_element(By.XPATH,'//*[@id="productSubtitle"]')
        b1=b.text
        c=a1+' '+b1
        e=driver.find_element(By.XPATH,'/html/body/div[2]/div/div[4]/div[1]/div[6]/div[7]/div[2]/div[2]/ul/li/span/span[1]/span/a/span[2]')
        price = e.text
        # price_1=price.split(' ')
        d=driver.find_element(By.XPATH,'//*[@id="landingImage"]')
        d1=d.get_attribute('src')

        b=driver.find_element(By.XPATH,'//*[@id="detailBulletsWrapper_feature_div"]')
        b1=b.text

        # Split the text by lines
        lines = [line.strip() for line in b1.split('\n') if line.strip()]

        # Create a dictionary to store the structured data
        product_details = {}

        # Extract the key-value pairs
        for line in lines[1:]:  # Skip the first line ("Product details")
            key, value = [part.strip() for part in line.split(':')]
            product_details[key] = value

        
        # Check if product_details is not empty
        data = {
            "title": c,
            "price": price,
            "image": d1,
            'product_details': product_details
        }
        dictionary.append(data)



    except:
        pass

    driver.execute_script("window.open('about:blank', 'tab2');")
    # driver.close()
    # time.sleep(2)
    driver.switch_to.window(driver.window_handles[-1])
    if len(data)==0:
        print('No data found or url not valid for '+str(i))

    else:
        print('Successfully collected data for '+str(i))

```

```python
dictionary
```

```python
json_string = json.dumps(dictionary,indent=4)

```

```python
print(json_string)
```

```python
file_name = "D:/relu/data.json"

# Write the JSON string to a JSON file
with open(file_name, "w") as json_file:
    json.dump(json_string, json_file)

print(f"Data has been written to {file_name}")
```
